### Performance comparation

```python
model_names_accurazy, accuracies = zip(*models_accurazy)
model_names_fscore, f_scores = zip(*models_fscore)

n_models = len(model_names_accurazy)
width = 0.35
ind = np.arange(n_models)

plt.figure(figsize=(10, 6))

# Plot bars for Accuracy and F-score
bars_accuracy = plt.bar(ind - width/2, accuracies, width, label='Accuracy', color='skyblue')
bars_fscore = plt.bar(ind + width/2, f_scores, width, label='F-score', color='salmon')

# Set labels and title
plt.xlabel('Models')
plt.ylabel('Performance')
plt.title('Model Accuracy vs F-score')
plt.ylim(0.0,1.0)
plt.xticks(ind, model_names_accurazy)  # Set model names on the x-axis
plt.legend()

# Add the values at the top of each bar
for bar in bars_accuracy:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

for bar in bars_fscore:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
